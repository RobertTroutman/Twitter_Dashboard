```python
df_etrade = df['eTrade'].dropna()
df_fidelity = df['Fidelity'].dropna()
df_Merrilledge = df['Merrilledge'].dropna()
df_robinhoodapp = df['robinhoodapp'].dropna()
df_tdameritrade = df['tdameritrade'].dropna()
```

```python

bossy_words = {"eTrade": '', "Fidelity": '', "Merrilledge": '', "robinhoodapp": '', "tdameritrade": ''}

df_list = [df_etrade, df_fidelity, df_Merrilledge, df_robinhoodapp, df_tdameritrade]

for key, df in zip(bossy_words, df_list):
    vectorizer = CountVectorizer(stop_words="english")
    data = vectorizer.fit_transform(df)
    words = pd.DataFrame(
            list(zip(vectorizer.get_feature_names(), np.ravel(data.sum(axis=0)))),
            columns=["Word", "Frequency"],)
    # Order the DataFrame by word frequency in descending order
    words_frequency = words.sort_values(by=["Frequency"], ascending=False)
    bossy_words[key] = words_frequency
```

```python
def process_text(doc):
    #sw = set(stopwords.words('english'))
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', doc)
    words = word_tokenize(re_clean)
    lem = [lemmatizer.lemmatize(word) for word in words]
    output = [word.lower() for word in lem] #if word.lower() not in sw]
    return ' '.join(output)
```

```python
cloud_df = pd.DataFrame(tweets_df)

cloud_df_etrade = cloud_df['eTrade'].dropna()
cloud_df_fidelity = cloud_df['Fidelity'].dropna()
cloud_df_Merrilledge = cloud_df['Merrilledge'].dropna()
cloud_df_robinhoodapp = cloud_df['robinhoodapp'].dropna()
cloud_df_tdameritrade = cloud_df['tdameritrade'].dropna()
```

```python
def generate_cloud(cloud_df_etrade):
    big_string = ' '.join(cloud_df_etrade)
    input_text = process_text(big_string)
    wc = WordCloud().generate(input_text)
    return plt.imshow(wc)
```
